#### 1.1. Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url_1 = 'https://github.com/trending/python?since=daily'
````

````python
response = requests.get(url_1)
````

````python
python_soup = BeautifulSoup(response.content)
````

````python
# your code here
repos = python_soup.find_all('h1', attrs={'class' : 'h3 lh-condensed'})
repos
````

````python
count = 0
for line in repos[count].find_all('span'):
    count = count + 1
    print(line.text.strip())
````

````python
repos_list = [] #define an empty list
for line in repos: #create a loop to go troguh all tags in the soup
    repos_list.append(line.text.strip())#append all names to a list and clean with strip()
print(repos_list) #creating a clean list of the names
len(repos_list)
````

````python
counter = 0
while counter < 10:
    for line in repos: #create a loop to go troguh all tags in the soup
        repos_list.append(line.text.strip())#append all names to a list and clean with strip()
        print(repos_list) #creating a clean list of the names
        counter = counter + 1
````
